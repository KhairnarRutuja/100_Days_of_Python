## Categorical Plot
A categorical plot is a type of visualization used to represent categorical data. Categorical data can include variables that can take on a limited and fixed number of possible values.

### 1. Strip Plot
A strip plot is a type of scatter plot where individual data points are plotted along a single axis, typically used to show the distribution of a dataset. It is useful for visualizing small datasets or overlapping points in a categorical variable.

```python
sns.stripplot(data=tips, x= 'day', y = 'total_bill')
```

```python
# jitter --> True (add small random noise to the position of data points, preventing excessive overlap and making the distribution clearer.)
sns.stripplot(data=tips, x= 'day', y = 'total_bill',jitter=False)
```

### 2. Cat Plot
A Cat Plot (Categorical Plot) is a type of visualization used to display data across categorical variables. It is a figure-level function in Seaborn that allows for different types of categorical plots like strip plots, box plots, violin plots, bar plots, and more.

```python
sns.catplot(data = tips, x = "day", y = "total_bill",kind = "strip")
```

```python
sns.catplot(data = tips, x = "day", y = "total_bill",kind = "strip",jitter=0.2,hue ="sex")
```

### 3. Swarmplot
A swarm plot is a type of scatter plot where data points are adjusted (avoiding overlap) so that each point remains visible. It is useful for visualizing the distribution of a categorical variable while showing each data point individually.

```python
sns.catplot(data=tips, x="day", y= "total_bill", kind ="swarm",hue = "sex")
```

```python
sns.swarmplot(data=tips, x="day", y= "total_bill", hue = "sex")
```

### 4. Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile [Q1], median, third quartile [Q3] and maximum). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
sns.boxplot(data = tips, x = "day", y = "total_bill")
```

```python
sns.catplot(data = tips, x = "day", y = "total_bill",kind="box",hue = "sex")
```

```python
sns.boxplot(data = tips, x = "day", y = "total_bill",hue="sex")
```

```python
sns.boxplot(data=tips,y ="total_bill")
```

### 5. Violinplot = (Boxplot + KDEPlot)
A violin plot is a combination of a box plot and a kernel density plot, used to visualize the distribution of numerical data across different categories.

```python
sns.violinplot(data = tips, x= "day",y ="total_bill",hue = "sex")
```

```python
sns.catplot(data = tips, x= "day",y ="total_bill",hue = "sex",kind = "violin")
```

```python
sns.catplot(data = tips, x= "day",y ="total_bill",hue = "sex",kind = "violin",split=True)
```

### 6. Bar Plot
A bar plot is used to show the comparison of numerical values across categorical variables. It typically represents aggregated values (e.g., mean, sum, count) for each category with bars.

```python
# ci=None: Removes confidence intervals (error bars)
sns.barplot(data=tips, x = "sex", y = "total_bill",hue="smoker",ci=None)
```

```python
# estimator: Changes the aggregation function (e.g., mean, sum, median, etc.)
import numpy as np
sns.barplot(data=tips, x = "sex", y = "total_bill",hue="smoker",estimator=np.std)
```

```python
sns.barplot(data=tips, x = "sex", y = "total_bill",hue="smoker",estimator=np.median)
```

```python
sns.barplot(data=tips, x = "sex", y = "total_bill",hue="smoker",estimator=np.min)
```

### 7. Point Plot
A point plot is similar to a bar plot, but instead of bars, it uses points connected by lines to show the trend of a numerical variable across different categories.

```python
sns.pointplot(data=tips, x = "sex", y = "total_bill",estimator=np.mean,hue='smoker')
```

### 8. Countplot
A count plot is a type of bar plot that shows the count (frequency) of occurrences for each category. It is useful when analyzing categorical data.

```python
sns.countplot(data=tips, x = "sex", hue="day")
```

```python
# faceting using catplot
sns.catplot(data=tips,x="sex",y="total_bill",col="smoker",kind="box",hue="sex",row="time")
```

## Regression Plots
A regression plot is used to visualize the relationship between two numerical variables while also displaying a linear regression line (trend line). It helps in identifying trends, patterns, and correlations in data.

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

### 1. RegPlot
The regplot function in Seaborn is used to plot a scatter plot with a regression line, helping visualize the relationship between two numerical variables.

```python
# hue parameter is not availabel
sns.regplot(data=tips,x = "total_bill",y ="tip")
```

```python
sns.lmplot(data=tips,x = "total_bill",y ="tip",hue="sex")
```

### 2. Residplot
A residual plot (residplot) is used to check the goodness of fit of a regression model. It plots residuals (errors) between observed and predicted values to help determine if a linear regression model is appropriate.

```python
sns.residplot(data=tips,x="total_bill",y="tip")
```

#### A Second way to plot facet plots-->FacetGrid

```python
sns.catplot(data=tips, x='sex', y= 'total_bill', kind= 'violin', col = 'day',row='time',hue='sex')
```

```python
g = sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.boxplot, 'sex', 'total_bill')
```

## Plotting Pairwise Relationship (PairGrid Vs Pairplot)

```python
iris
```

### Pairplot
pairplot is a function in Seaborn that creates a grid of scatterplots (or histograms) to visualize pairwise relationships between numerical variables in a dataset.

```python
sns.pairplot(iris,hue="species")
```

### 2. PairGrid
PairGrid is a more flexible and customizable version of sns.pairplot. While pairplot automatically creates a grid of scatterplots for pairwise comparisons of numerical variables, PairGrid allows you to customize each part of the grid separately.

```python
g = sns.PairGrid(data = iris,hue = "species")
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris,hue = "species")
g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris,hue = "species")
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

### JointGrid and JointPlot
Both JointGrid and jointplot are functions from the Seaborn library in Python, used for visualizing the relationship between two variables along with their marginal distributions. However, they have different levels of flexibility and ease of use.

```python
sns.jointplot(data=tips,x="total_bill",y = 'tip',kind = 'hist',hue = "sex" )
```

```python
g = sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.kdeplot,sns.violinplot)
```
