# Pandas DataFrame
Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data, rows, and columns.
Pandas DataFrame will be created by loading the datasets from existing storage, storage can be SQL Database, CSV file, and Excel file. Pandas DataFrame can be created from the lists, dictionary, and from a list of dictionary etc.

```python
import numpy as np
import pandas as pd
```

### Creating DataFrame
To create DataFrame from dict of narray/list, all the narray must be of same length. If index is passed then the length index should be equal to the length of arrays. If no index is passed, then by default, index will be range(n) where n is the array length.

```python
# create empty dataframe
df = pd.DataFrame()
print(df)
print(type(df))
```

```python
# create dataframe using list
student_data = [
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]

df1 = pd.DataFrame(student_data, columns = ['iq','marks','package'])
df1
```

```python
# using dictionary

student_dict = {
    'name':['A','B','C','D'],
    'iq':[100,90,120,80],
    'marks':[80,70,100,50],
    'package':[10,7,14,2]
}
df3 = pd.DataFrame(student_dict)
df3.set_index('name',inplace=True)
df3
```

### Selecting rows from a DataFrame
Indexing operator is used to refer to the square brackets following an object. The .loc and .iloc indexers also use the indexing operator to make selections.

**1. iloc:**
This method is integer position-based, meaning it selects data based on the numerical position of rows and columns, similar to standard Python list indexing.

**2. loc:**
This method is label-based, meaning it selects data based on the labels of rows and columns. It allows for selection using row and column labels, and it can also handle boolean arrays for conditional filtering.

```python
# single row
df4.iloc[0]
```

```python
df4.iloc[5]
```

```python
# multiple column
df4.iloc[0:5]
```

```python
df4.iloc[5:16:2]
```

```python
# fancy indexing
df4.iloc[[0,4,5]]
```

```python
# loc
df3.loc['A']
```

```python
df3.loc['A':'C']
```

```python
df3
```

### Filtering a DataFrame

```python
df5.head()
```

```python
# find all the final winners
mask = df5['MatchNumber'] == 'Final'
new_df = df5[mask]
new_df[['Season','WinningTeam']]
```

```python
df5[df5['MatchNumber'] == 'Final'][['Season','WinningTeam']]
```

```python
# how many super over finishes have occured
df5[df5['SuperOver'] == 'Y'].shape[0]
```

```python
# how many matches has csk won in kolkats
df5[(df5['City'] == 'Kolkata') & (df5['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# toss winner is match winner in percentage
df5[df5['TossWinner'] == df5['WinningTeam']].shape[0]/df5.shape[0]*100
```

```python
# movies with rating higher than 8 and votes > 10000
df4[(df4['imdb_rating'] > 8) & (df4['imdb_votes'] > 10000)].shape[0]
```

```python
# Action movies with rating higher than 7.5
mask1 = df4['genres'].str.split('|').apply(lambda x: 'Action' in x)
mask2 = df4['imdb_rating'] > 7.5
df4[mask1 & mask2]
```

```python
mask1 = df4['genres'].str.contains('Action')
mask2 = df4['imdb_rating'] > 7.5
df4[mask1 & mask2]
```
