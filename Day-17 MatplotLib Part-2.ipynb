## Colored Scatterplots

```python
Df = pd.read_csv('iris.csv')
Df.head()
```

```python
Df['Species'] = Df['Species'].replace({'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2})
Df.sample(5)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(Df['SepalLengthCm'],Df['PetalLengthCm'],c=Df['Species'],cmap='jet',alpha=0.7)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.colorbar()
plt.show()
```

## Annotations

```python
Df1 = pd.read_csv("batter.csv")
Df1.head(5)
```

```python
sample_df = Df1.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.figure(figsize=(18,6))
plt.scatter(sample_df["avg"],sample_df["strike_rate"],s=sample_df["runs"])
for i in range(sample_df.shape[0]):
    plt.text(sample_df["avg"].values[i],sample_df['strike_rate'].values[i],sample_df["batter"].values[i])
```

```python
plt.figure(figsize=(18,6))
plt.scatter(sample_df["avg"],sample_df["strike_rate"],s=sample_df["runs"])

plt.axhline(130,color = 'red')
plt.axhline(140,color = 'green')
plt.axvline(30,color = 'red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df["avg"].values[i],sample_df['strike_rate'].values[i],sample_df["batter"].values[i])
```

## 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize = (12,8))

ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize = (12,8))

ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(yy)

fig = plt.figure(figsize = (12,8))

ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## Contour PLot

```python
fig = plt.figure(figsize = (12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize = (12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize = (12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## Heatmap

```python
Df2 = pd.read_csv("IPL_Ball_by_Ball_2008_2022.csv")
Df2.head()
```

```python
temp_df = Df2[(Df2["ballnumber"].isin([1,2,3,4,5,6])) & (Df2["batsman_run"]==6)]
```

```python
grid = temp_df.pivot_table(index = "overs" , columns = "ballnumber",values = "batsman_run",aggfunc = "count")
```

```python
plt.figure(figsize=(5,5))
plt.imshow(grid)
plt.colorbar()
```
