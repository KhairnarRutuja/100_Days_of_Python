# Seaborn Package
Seaborn is a python visualization Library built on top of matplotlib and integrated with pandas. It provides a high-level interface for creating beautiful and informative statistical graphics with minimal code

#### Types of Functions
- Figure Level
- Axis Level

**1. Figure-Level Plots**

These are high-level interfaces that manage the overall figure, including multiple subplots.
They use the Figure object and automatically create a figure with subplots (axes).
These functions return a FacetGrid or PairGrid object, not a single Axes object.
use figure level functions if you want multiple subplots or facet grids (eg., different categories in seperate plots)

**category Figure_level function**

1. Relational Plots    -   relPlot
2. Categorical Plots   -   catplot
3. Distribution plots  -   displot
4. Regression plots    -   lmplot
5. Pairwise plots      -   pairplot



**2. Axes-Level Plots**

These are low-level interfaces that create a single plot inside an Axes object.
They modify a single Axes inside a Figure.
use Axes level function if you need fine control over a single plot inside a matplotlib figure

**category Axes_level function**

1. Relational Plots    -   Scatterplot,Lineplot
2. Categorical Plots   -   Boxplot,violinplot,barplot,countplot,stripplot,swarmplot
3. Distribution plots  -   histplot,kdeplot
4. Regression plots    -   regplot
5. Matrix plots        -   heatplot

## 1. Relational Plot
Relational plots are used to visualize relationship between two or more variables in a dataset They help in understanding how data points are distributed and how variables interact with each otheersis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### 1.  Scatter Plot 
A scatter plot is used to visualize the relationship between two numerical variables.
when analyzing correlations or trends between two numerical variable

```python
# style --> different markers(symbols)
# sizee --> changes point size 
sns.scatterplot(data = tips , x = "total_bill", y = "tip", hue = "sex", style = "time",size = "size")
```

### Relplot
relplot is a powerfull function in seaborn used to create relational plots (scatter plots and line plots) with additional flexibility. It acts as a figure level interface for scatterplot and lineplot allowing easy faceting(creating multiple subplots based on categories).

### When to use relplot
- When we need flexibility(scatter or line plots).
- when we want to facet the data using row and col.
- when you need better plot styling using hue,size,and styple.

```python
# plot scatter plot using relplot
# kind --> type of plot (scatter or line)
sns.relplot(data = tips, x = "total_bill", y = "tip", kind = "scatter", hue = "sex" , style = "time",size = "size" )
```

### 2. Line Plot
A line plot is used to show trends over time or relationship between two numerical variables.
when wroking with time series data or tracking changes over time

```python
gap = px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
sns.lineplot(data=temp_df, x = 'year', y = 'lifeExp')
```

```python
# plot line plot using relplot
sns.relplot(data = temp_df, x = "year", y = "lifeExp", kind = "line")
```

```python
temp_df= gap[gap["country"].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.relplot(kind = 'line', data = temp_df, x = 'year', y = 'lifeExp', hue = 'country', style = 'continent',size = 'continent')
```

```python
sns.lineplot(data = temp_df, x = 'year', y = 'lifeExp', hue = 'country')
```

### Facet Plot
Facet plots also called small multiples or faceted visualization are useful for visualizing data across multiple subsets.
seaborn provides facetgrid and figure level functions like relplot, catplot, displot, to create facet plots easily.

```python
# col --> create seperate plots for each categories vertical
# row --> create seperate plots for each categories horizontal
sns.relplot(data = tips , x = 'total_bill', y = 'tip' , kind = "scatter" , col = 'smoker',row = 'sex')
```

```python
# column Wrap
sns.relplot(data = gap , x = 'lifeExp', y = 'gdpPercap', kind = 'scatter', col = 'year', col_wrap = 3)
```

## 2. Distribution Plots

Distribution plots help visualize the spread and shape of a dataset. seaborn provides multiple way to plot distributions displot,histplot,kdeplot

### 1. Histogram plot
A Histogram represents the frequency distribution of numerical data by dividing values into bins and counting occurrences in each bin

```python
sns.histplot(data = tips , x = 'total_bill')
```

### 2. Displot
Displot is a figure level function that can create histograms and KDEs with faceting support 

```python
sns.displot(data = tips, x = 'total_bill', kind = 'hist')
```

```python
# Bin Parameter (number of bins (default is auto-selected)
sns.displot(data = tips, x = 'total_bill', kind = 'hist', bins = 20)
```

```python
# Its also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables

# countplot
sns.displot(data = tips, x = "day", kind = "hist")
```

```python
# hue Parameter (groups data by a categorical column)
sns.displot(data = tips , x = 'tip',kind = 'hist', hue = 'sex')
```

```python
# elements = "bars" (default), "step" (line histogram) ,"poly" (polygon histogram)
sns.displot(data = tips, x = 'tip', kind = 'hist', hue = 'sex', element = 'step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
# Faceting using col and row -- not work on histplot function
sns.displot(data = titanic , x = 'age' , kind = 'hist', element = 'step',col = 'sex')
```

### 3. kdeplot
KDE (Kernel Density Estimation) is a smoothed version of a histogram estimating the probability density function (PDF)

```python
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data = tips, x = 'total_bill' , kind = 'kde')
```

```python
# fill (if true fills the area under the KDE curve)
sns.displot(data = tips, x = 'total_bill' , kind = 'kde',hue = 'sex',fill = True)
```

### Rugplot
A rug plot is a simple visualization that adds small ticks along the x- axis to show individual data points It is usefull for understanding data distribution especially when used alongside histogram or kDE Plot

```python
# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(data=tips, x='total_bill', y='tip')
```

```python
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

## Matrix Plot
Matrix plots are used to visualize relationship between multiple numerical variables in structured way 
- Heatmap
- Clustermap

### 1. Heatmap
A heatmap display values in matrix format where color intensity represents magnitude

```python
temp_df = gap.pivot(index = "country", columns = "year", values = 'lifeExp')
temp_df
```

```python
sns.heatmap(temp_df)
```

```python
# Annot (True--> Display numbers on each cell)
# camp (sets the color scheme)
plt.figure(figsize=(15,15))
temp_df = gap[gap['continent'] == 'Europe'].pivot(index = 'country',columns='year',values='lifeExp')
sns.heatmap(temp_df,annot= True,linewidth=0.5,cmap='summer')
```

### 2. ClusterMap
A cluster map clusters similar data points together using hierarchnical clustering

```python
iris = px.data.iris()
iris.head()
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
