## 17. apply
Using this we can apply a function to every row in the given dataframe. 

```python
# apply on series
temp = pd.Series([10,20,30,40,50])
temp
```

```python
def sigmoid(value):
    return 1/1+np.exp(-value)
```

```python
temp.apply(sigmoid)
```

```python
# apply on dataframe
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)
points_df
```

```python
def euclidean(row):
    pt_A = row['1st point']
    pt_B = row['2nd point']
    return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
```

## 18. isin
Checks whether each element in the DataFrame is contained in a specified collection (iterable, Series, DataFrame, or dictionary), returning a DataFrame of booleans.

```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
mask = df.isin([2, 5])
print(mask)
```

## 19. corr 
Computes pairwise correlation of columns, excluding NA/null values.

```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
correlation_matrix = df.corr()
print(correlation_matrix)
```

## 20. nlargest and nsmallest
Retrieve the first n rows ordered by columns in descending (nlargest) or ascending (nsmallest) order.

```python
df = pd.DataFrame({'A': [10, 20, 20, 30], 'B': [4, 3, 2, 1]})
top_rows = df.nlargest(2, 'A')
bottom_rows = df.nsmallest(2, 'A')
print(top_rows)
```

```python
print(bottom_rows)
```
