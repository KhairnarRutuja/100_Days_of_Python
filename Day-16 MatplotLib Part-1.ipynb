# Matplotlib Package
Matplotlib is a powerful and widely used plotting library in Python that allows you to create a variety of visualizations, including line plots, bar charts, histograms, scatter plots, and more.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2D Line 

- Bivariate Analysis
- combination - (Categorical,Numerical),(Numerical,Numerical)
- use Case - Time Series Data

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
plt.figure(figsize=(5,3))  # Set figure size
plt.plot(year, price, marker='o', linestyle='-', color='b', label="Price Trend")

plt.xlabel("Year")
plt.ylabel("Price")
plt.title("Price Trend Over Years")
plt.grid(True)  # Add grid
plt.legend()  # Show legend
plt.show()
```

```python
# pandas dataframe
df = pd.read_csv("sharma-kohli.csv")
df
```

```python
plt.figure(figsize=(5,3))
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
```

```python
# labels title
plt.figure(figsize=(5,3))
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
plt.title('Rohit sharma vs Virat Kohali Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
# colors(hex) and line(width and style)) and marker(size)
plt.figure(figsize=(5,3))
plt.plot(df['index'],df['V Kohli'],color='green')
plt.plot(df['index'],df['RG Sharma'],color='black')
plt.title('Rohit sharma vs Virat Kohali Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
# colors(hex) and line(width and style)) and marker(size)
plt.figure(figsize=(5,3))
plt.plot(df['index'],df['V Kohli'],color='green',linestyle='dashed',linewidth = 2)
plt.plot(df['index'],df['RG Sharma'],color='black',linestyle= 'dotted',linewidth = 1)
plt.title('Rohit sharma vs Virat Kohali Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
plt.figure(figsize=(5,3))
plt.plot(df['index'],df['V Kohli'],color='green',linestyle='dashed',linewidth = 2,marker = 'D',markersize = 5)
plt.plot(df['index'],df['RG Sharma'],color='black',linestyle= 'dotted',linewidth = 1,marker = 'o',markersize = 5)
plt.title('Rohit sharma vs Virat Kohali Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
plt.figure(figsize=(5,3))
plt.plot(df['index'],df['V Kohli'],color='green',linestyle='dashed',linewidth = 2,marker = 'D',markersize = 5, label = 'virat')
plt.plot(df['index'],df['RG Sharma'],color='black',linestyle= 'dotted',linewidth = 1,marker = 'o',markersize = 5,label = "Rohot")
plt.title('Rohit sharma vs Virat Kohali Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
plt.legend(loc='best')
```
